```python
# Load jsonl file with all the data
import json
import pandas as pd

with open('../src/data/final/data_prepared.jsonl') as f:
    data = [json.loads(line) for line in f]

df = pd.DataFrame(data)
df.head()
```

```python
# For every prompt strip the whitespaces in the beginning and end
df['prompt'] = df['prompt'].str.strip()
df['completion'] = df['completion'].str.strip()
df.head()
```

```python
# For every prompt add Question: at the beginning
df['prompt'] = 'Question to db: ' + df['prompt']
df.head()
```

```python
# For every completion add Answer: at the beginning
df['completion'] = 'Answer: ' + df['completion']
df.head()
```

```python
# Export to jsonl
# df as utf8
df.to_json('../src/data/final/data_prepared.jsonl', orient='records', lines=True, force_ascii=False)
```
